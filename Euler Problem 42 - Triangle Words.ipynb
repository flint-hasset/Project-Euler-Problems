#### 42. The nth term of the sequence of triangle numbers is given by, tn = Â½n(n+1); so the first ten triangle numbers are:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

By converting each letter in a word to a number corresponding to its alphabetical position and adding these values we form a word value. For example, the word value for SKY is 19 + 11 + 25 = 55 = t10. If the word value is a triangle number then we shall call the word a triangle word.

Using words.txt (right click and 'Save Link/Target As...'), a 16K text file containing nearly two-thousand common English words, how many are triangle words?

```python
# Create list of triangle numbers
triangle_list = []
for n in range(0,100):
    triangle_list.extend([0.5*n*(n+1)])

# Then create word_scores and check

word_list = open("Z:\Documents\Projects\Euler Problems\p042_words.txt").read().split(',')

j=0
for word in word_list:
    word_list[j]=word[1:-1]
    j+=1
   
    
lets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
b_value = lets.find('B')+1
triangle_count = 0

for word in word_list:
    word_score = 0
    for char in word:
        t_val = lets.find(char)+1
        word_score = word_score + t_val
    if word_score in triangle_list:
        triangle_count = triangle_count + 1

print(triangle_count)

    
```
